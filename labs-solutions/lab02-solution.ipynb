## Part 4

I clean up the date here. There are other ways to do this, but I like seeing how we can go from **text to an actual date**. Once Python and `pandas` knows that something is a date, you can use date-specific functions on the variable. Just like in Excel.

```python
kc['date'] = kc['date'].str.replace('T000000','')
```

```python
print(kc[['date', 'price', 'sqft_living']])
```

```python
kc[['date', 'price', 'sqft_living']]
```

## Part 5

Now, I'm going to format my date and check the variable types in the data set.

```python
kc['date'] = pd.to_datetime(kc['date'], format='%Y%m%d')

kc.info()
```

```python
kc['year'] = pd.DatetimeIndex(kc['date']).year
kc['month'] = pd.DatetimeIndex(kc['date']).month
```

## Part 7

```python
print(kc['date'].min())
print(kc['date'].max())
```

```python
kc['ym'] = kc['date'].dt.strftime('%Y-%m')
```

```python
ym_groups = kc.groupby('ym')
print(ym_groups.ngroups)
print(ym_groups.size())
```

```python
print(ym_groups['price'].mean())
```

```python
kc.groupby('ym')['price'].mean()
```

```python
kc.groupby('ym')['price'].agg([np.mean, np.median, np.std]) 
```

I'm using that `.style` again here. Looks pretty good!

```python
kc.groupby(['year', 'month'])['price'].agg([np.mean, np.median, np.std]).style.format("{:,.2f}") 
```
