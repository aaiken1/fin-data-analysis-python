```python
import numpy as np
import pandas as pd
import datetime as datetime
```

```python
kc = pd.read_csv('https://raw.githubusercontent.com/aaiken1/fin-data-analysis-python/main/data/kc_house_data.csv',
                      index_col=0)
```

```python
kc.info()
```

```python
kc.index
```

```python
kc['date'] = kc['date'].str.replace('T000000','')
```

```python
print(kc[['date', 'price', 'sqft_living']])
```

```python
kc['date'] = pd.to_datetime(kc['date'], format='%Y%m%d')

kc.info()
```

```python
kc['year'] = pd.DatetimeIndex(kc['date']).year
kc['month'] = pd.DatetimeIndex(kc['date']).month
```

```python
print(kc.shape)
print(kc.columns)
```

```python
kc.describe()
```

```python
kc[['price', 'bedrooms', 'bathrooms', 'sqft_living']].describe()
```

```python
pd.options.display.float_format = '{:,.2f}'.format
kc[['price', 'bedrooms', 'bathrooms', 'sqft_living']].describe()
```

```python
kc.isna().sum()
```

```python
kc[['view', 'condition', 'grade']].nunique()
```

```python
kc.query('price > 5000000')
```

```python
kc.query('price > 1000000' and 'bedrooms >= 5')
```

```python
print(kc['date'].min())
print(kc['date'].max())
```

```python
kc['ym'] = kc['date'].dt.strftime('%Y-%m')
```

```python
ym_groups = kc.groupby('ym')
print(ym_groups.ngroups)
print(ym_groups.size())
```

```python
print(ym_groups['price'].mean())
```

```python
kc.groupby('ym')['price'].mean()
```

```python
kc.groupby('ym')['price'].agg([np.mean, np.median, np.std]) 
```

```python
pd.options.display.float_format = '${:,.0f}'.format
kc.groupby(['year', 'month'])['price'].agg([np.mean, np.median, np.std]) 
```
